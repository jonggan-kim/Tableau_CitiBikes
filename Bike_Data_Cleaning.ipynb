```python
#Import dependencies
import path
import pandas as pd
# import glob
```

```python
df = pd.read_csv("AllBikesData.csv")
df.head()
```

```python
#Update DataFrame with starttime as index
df = df.set_index(pd.DatetimeIndex(df['starttime']))

df['Hour'] = df.index.hour
df['Month'] = df.index.month

df.head()
```

```python
#Export frame to CSV
df.to_csv('AllBikesDataUpdated.csv')
```
